```python
# Membandingkan top produksi komoditas Indonesia pada negara tetangga
countries = ['Indonesia', 'India', 'China, mainland', 'Singapore', 'Malaysia']

df_selected = df[
    (df["Area"].isin(countries)) &
    (df["Element"] == "Production")
]

top_items_per_country = (
    df_selected.groupby(["Area", "Item"])["Value"]
    .sum()
    .reset_index()
    .sort_values(["Area", "Value"], ascending=[True, False])
)

top_items_dict = top_items_per_country.groupby("Area").head(5).groupby("Area")["Item"].unique().to_dict()

df_plot = df_selected[
    df_selected.apply(lambda row: row["Item"] in top_items_dict[row["Area"]], axis=1)
]

df_tren_all = df_plot.groupby(["Area", "Year", "Item"])["Value"].sum().reset_index()

fig, axes = plt.subplots(2, 3, figsize=(16, 12), sharex=True)
axes = axes.flatten()

for i, country in enumerate(countries):
    ax = axes[i]
    data = df_tren_all[df_tren_all["Area"] == country]
    sns.lineplot(data=data, x="Year", y="Value", hue="Item", marker="o", ax=ax)
    ax.set_title(f"Produksi Komoditas di {country}")
    ax.set_xlabel("Tahun")
    ax.set_ylabel("Produksi (ton)")
    ax.grid(True)
    ax.legend(title="Komoditas", loc='best')

plt.tight_layout()
plt.show()
```

```python
df_idn = df[(df["Area"] == "Indonesia") & (df["Element"] == "Production")]
pivot_prod = df_idn.pivot_table(index="Year", columns="Item", values="Value", aggfunc="sum")

recent_years = pivot_prod.loc[2010:2020]

def compute_cagr(start, end, periods):
    return (end / start) ** (1 / periods) - 1

cagr_dict = {}
for item in recent_years.columns:
    series = recent_years[item].dropna()
    if len(series) >= 2 and series.iloc[0] > 0:
        cagr = compute_cagr(series.iloc[0], series.iloc[-1], len(series) - 1)
        cagr_dict[item] = cagr

cagr_df = pd.DataFrame.from_dict(cagr_dict, orient="index", columns=["CAGR"]).sort_values(by="CAGR", ascending=False)

predictions = {}
for item, row in cagr_df.iterrows():
    last_val = recent_years[item].dropna().iloc[-1]
    cagr = row["CAGR"]
    pred_2025 = last_val * ((1 + cagr) ** 5)
    pred_2029 = last_val * ((1 + cagr) ** 9)
    predictions[item] = {
        "2020": last_val,
        "2025_pred": pred_2025,
        "2029_pred": pred_2029,
        "CAGR": cagr
    }

pred_df = pd.DataFrame(predictions).T.sort_values(by="2029_pred", ascending=False)

top5 = pred_df.head(5)
print(top5)

top5[["2020", "2025_pred", "2029_pred"]].plot(kind="bar", figsize=(12,6))
plt.title("Prediksi Produksi 5 Komoditas Unggulan Indonesia (2020–2029)")
plt.ylabel("Produksi (ton)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
